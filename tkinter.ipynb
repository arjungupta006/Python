```python
from tkinter import * 
import sqlite3
root = Tk()
root.title("Database Connection")
root.geometry("400x200")
def connect_db():
    conn = sqlite3.connect('example.db')
    print("Database connected successfully!")
    conn.close()
connect_button = Button(root, text="Connect to Database", command=connect_db)
connect_button.pack(pady=20)

root.mainloop()
```
